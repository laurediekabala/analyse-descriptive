## Décomposition du KHI-2

```python
#KHI-2 du tableau de contingence
from scipy.stats import chi2_contingency
res = chi2_contingency(df,correction=False)
res
```

```python
#info dispo - khi2
res.statistic
```

```python
#décomposée comme suit
contribKhi2 = ((df.values - res.expected_freq)**2)/res.expected_freq

#vérification de la somme
np.sum(contribKhi2)
```

```python
#en fraction de l'info disponible
frac_contrib = contribKhi2/res.statistic
df_contrib = pd.DataFrame(frac_contrib,index=df.index,columns=df.columns)
df_contrib
```

```python

sn.heatmap(df_contrib,vmin=0,vmax=np.max(frac_contrib),cmap="Greens")
```

### Résidus standardisés

```python
#résidus standardisés
res_std = (df.values-res.expected_freq)/np.sqrt(res.expected_freq)
df_res_std = pd.DataFrame(res_std,index=df.index,columns=df.columns)
df_res_std
```

```python
#sous forme de heatmap
sn.heatmap(df_res_std,center=0.0,cmap=sn.diverging_palette(10,250,as_cmap=True))
```
